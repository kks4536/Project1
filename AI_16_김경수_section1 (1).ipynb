```python
for i, k in enumerate(data['NA_Sales']):
  if 'M' in data['NA_Sales'][i]:
    New_Sales = float(k.replace('M', ''))*(1000000)
    data['NA_Sales'][i] = New_Sales
  else:
    New_Sales = float(k.replace('K', ''))*(1000)
    data['NA_Sales'][i] = New_Sales


for i, k in enumerate(data['EU_Sales']):
  if 'M' in data['EU_Sales'][i]:
    New_Sales = float(k.replace('M', ''))*(1000000)
    data['EU_Sales'][i] = New_Sales
  else:
    New_Sales = float(k.replace('K', ''))*(1000)
    data['EU_Sales'][i] = New_Sales


for i, k in enumerate(data['JP_Sales']):
  if 'M' in data['JP_Sales'][i]:
    New_Sales = float(k.replace('M', ''))*(1000000)
    data['JP_Sales'][i] = New_Sales
  else:
    New_Sales = float(k.replace('K', ''))*(1000)
    data['JP_Sales'][i] = New_Sales


for i, k in enumerate(data['Other_Sales']):
  if 'M' in data['Other_Sales'][i]:
    New_Sales = float(k.replace('M', ''))*(1000000)
    data['Other_Sales'][i] = New_Sales
  else:
    New_Sales = float(k.replace('K', ''))*(1000)
    data['Other_Sales'][i] = New_Sales


    
data.loc[:,('NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales')] = data.loc[:,('NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales')].astype('float64')
data

```

```python
df_isnull = data[data.Year.isnull() | data.Genre.isnull() | data.Publisher.isnull()]
df_isnull
```

```python
data = data.drop(df_isnull.index)
data
```

```python
data.Year = data.Year.astype('int64')
```

```python
data = data.drop('Unnamed: 0', axis = 1)
data = data.reset_index(drop=True)
data
```

```python
Total_Sales = data.NA_Sales + data.EU_Sales + data.JP_Sales + data.Other_Sales
```

```python
data['Total_Sales'] = Total_Sales
data
```

```python
data = data.sort_values(by = 'Total_Sales', ascending = False)
data
```

```python
data = data.reset_index(drop = True)
```
